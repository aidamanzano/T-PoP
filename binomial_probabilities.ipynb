```python
import numpy as np
import matplotlib.pyplot as plt

plots_path = '/Users/amm3117/Desktop/Proof-of-Location/Proof-of-Location/probabilities/'
fig, ax = plt.subplots(6, 2, figsize = (20, 40))
ax = ax.ravel()

n = 6 #number of trials
argmax_probabilities = []
probability_of_picking_lying_car = []
for p in range(0, 110, 10):
    
    i = int(p/10)
    print(i)
    p = p/100 #probability of sampling a lying car
    print(p)
    probability_of_picking_lying_car.append(p)
    q = 1 - p
    lying_cars = np.array((0, 1, 2, 3, 4, 5, 6))

    probabilities = []

    for X in range(0, 7): #X is the number of cars sampled that are lying

        P_X = (np.math.factorial(n)/((np.math.factorial(n - X)) * (np.math.factorial(X)))) * ((p)**(X)) * ((q)**(n-X))
        probabilities.append(P_X)

    p_max = max(probabilities)
    argmax_probabilities.append(p_max)

    ax[i].stem(lying_cars, probabilities)
    ax[i].set(xlabel='lying cars picked', ylabel='probability')
    ax[i].set_title('probability of picking lying car ' + str(p))



plt.stem(probability_of_picking_lying_car, argmax_probabilities)
plt.title('Maximum probability for each possible p value')
plt.xlabel('Maximum probability')
plt.ylabel('Probability p of picking a lying car')
plt.show()

fig = fig.get_figure()
fig.savefig(plots_path + 'lying_cars.png')
plt.show()

#TODO: 
""" Change plots to discrete probabilities (stem plots)
the non-identical behaviour of the probabilities
draw 
 """
```
